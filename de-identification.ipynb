```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
#check raw images and load images list

input_images = '/root/KJS/de-identification/images/input'

for root, dirs, files in os.walk(input_images):
    print('The number of raw images :', len(files))
    for fname in files:
        if not fname.endswith('.jpg'):
            continue
        full_fname = os.path.join(root,fname)
        image = img.imread(full_fname)
    print('example image')
    plt.imshow(image)
    plt.show()
```
